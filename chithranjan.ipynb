```python
# ============================================
# Smart Weather Accident Risk Predictor
# Google Colab Web App (Guaranteed Output)
# ============================================

!pip install flask pandas numpy scikit-learn pyngrok

from flask import Flask, render_template_string, request
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from pyngrok import ngrok
import threading

# -------------------------------
# Train ML model (offline)
# -------------------------------
data = {
    'temperature': [30, 25, 20, 35, 18, 40, 22, 28],
    'humidity': [60, 80, 90, 50, 95, 40, 85, 70],
    'rain': [0, 1, 1, 0, 1, 0, 1, 0],
    'visibility': [10, 4, 3, 10, 2, 10, 4, 8],
    'risk': ['Low', 'High', 'High', 'Low', 'High', 'Low', 'High', 'Low']
}

df = pd.DataFrame(data)
le = LabelEncoder()
df['risk'] = le.fit_transform(df['risk'])

X = df.drop('risk', axis=1)
y = df['risk']

model = LogisticRegression()
model.fit(X, y)

print("âœ… Machine Learning model trained successfully")

# -------------------------------
# Flask App
# -------------------------------
app = Flask(__name__)

HTML = """
<!DOCTYPE html>
<html>
<head>
    <title>Smart Weather Accident Risk Predictor</title>
</head>
<body style="font-family:Arial;text-align:center">
    <h1>Smart Weather Accident Risk Predictor</h1>
    <form method="post">
        <input name="temperature" placeholder="Temperature (Â°C)" required><br><br>
        <input name="humidity" placeholder="Humidity (%)" required><br><br>
        <input name="visibility" placeholder="Visibility (km)" required><br><br>
        <select name="rain" required>
            <option value="">Is it Raining?</option>
            <option value="1">Yes</option>
            <option value="0">No</option>
        </select><br><br>
        <button type="submit">Predict Risk</button>
    </form>

    {% if result %}
        <h2>{{ result }}</h2>
    {% endif %}
</body>
</html>
"""

@app.route("/", methods=["GET", "POST"])
def home():
    result = None
    if request.method == "POST":
        temperature = float(request.form["temperature"])
        humidity = float(request.form["humidity"])
        visibility = float(request.form["visibility"])
        rain = int(request.form["rain"])

        features = np.array([[temperature, humidity, rain, visibility]])
        prediction = model.predict(features)

        result = "ðŸš¨ HIGH ACCIDENT RISK" if prediction[0] == 1 else "âœ… LOW ACCIDENT RISK"

    return render_template_string(HTML, result=result)

# -------------------------------
# Run Flask in background
# -------------------------------
def run_app():
    app.run(port=5000)

threading.Thread(target=run_app).start()

# -------------------------------
# Start ngrok tunnel
# -------------------------------
public_url = ngrok.connect(5000)
print("\nðŸŒ WEB APPLICATION LINK (CLICK THIS):")
print(public_url)
```
